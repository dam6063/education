```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
School = pd.read_csv('https://raw.githubusercontent.com/dam6063/education/master/data.csv', sep=',', header=0)
School.head(0)
```

```python
#Average pct_ccr and pct_glp by year for NC
School_summary = School.set_index('year').groupby('year')[['pct_ccr', 'pct_glp']].mean().reset_index()
School_summary
```

```python
#All Schools pct_glp vs. year Box Plots
with sns.axes_style(style='ticks'):
    g = sns.factorplot("year", "pct_glp", data=School, kind="box")
    g.set_axis_labels("Year", "Percent on Grade Level");
```

```python
#Plot of pct_glp vs. year
sns.set()
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [11,7]

plt.plot(School_summary['year'],School_summary['pct_glp'])
plt.title('Average percentage of Elemenatry Students at or above Grade Level in NC')
plt.xlabel('Year')
plt.ylabel('Percentage at or above Grade Level')
```

```python
#Plot of pct_ccr vs. year
sns.set()
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [11,7]

plt.plot(School_summary['year'],School_summary['pct_ccr'])
plt.title('Average percentage of Elemenatry Students CCR in NC')
plt.xlabel('Year')
plt.ylabel('Percentage of Students CCR')
```
