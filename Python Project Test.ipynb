```python
pub_mean = public.groupby(['County name', 'year'])['pct_glp'].agg('mean')
not_mean = notpublic.groupby(['County name', 'year'])['pct_glp'].agg('mean')
    
not_mean
```

```python
pub_all_mean=public.groupby('County name')['pct_glp'].agg('mean')
not_all_mean = notpublic.groupby('County name')['pct_glp'].agg('mean')
```

```python
not_all_mean
```

```python
pub_all_mean
```

```python
from scipy.stats import ttest_ind, ttest_rel, f_oneway
```

We will begin by testing for a difference in the means, using a t-test for difference in means:

```python
stat, p = ttest_ind(pub_all_mean, not_all_mean)
```

```python
print('Statistics=%.3f, p=%.5f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Same distributions (fail to reject H0)')
else:
    print('Different distributions (reject H0)')
```

We can also test this again using an anova test, but this is unnecessary since we only have two groups...

```python
stat1, p1 = f_oneway(pub_all_mean, not_all_mean)
```

```python
print('F Score=%.3f, p=%.5f' % (stat1, p1))
alpha = 0.05
if p1 > alpha:
    print('Same distributions (fail to reject H0)')
else:
    print('Different distributions (reject H0)')
```
